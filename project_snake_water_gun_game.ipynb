```python
#SNAKE WATER GUN GAME
#by Vinodhini Rajamanickam

print("WELCOME TO THE SNAKE-WATER-GUN GAME !!")

print("MODE : computer against player")
print("\n")
print("*********************************")


#import module
import random

#defining the class Swg_game
def Swg_game(computer, player):    

    if computer == player:              
        return None                            #if both computer's and player's values are samethe it is a tie
    

#checking for all the possibilities when computer chooses snake(s)
    elif computer == 's':
        if player == 'w':
            return False
        elif player== 'g':
            return True

#checking for all the possibilities when computer chooses water(w)        
    elif computer == 'w':
        if player == 'g':
            return False
        elif player== 's':
            return True

#checking for all the possibilities when computer chooses gun(g)
    elif computer == 'g':
        if player == 's':
            return False
        elif player == 'w':
            return True

#computer's turn  to choose the value
print("computer's turn : \nsnake(s) / water(w) / gun(g)")
computer_turn = random.randint(1,3)
if computer_turn ==1:
    computer ='s'
elif computer_turn == 2:
    computer = 'w'
elif computer_turn ==  3:
    computer = 'g'

print("the computer has chosen")
print("\n")

#the player's turn to choose the value and get the value as the input 
player = input("player's turn : select any one:\nsnake(s) / water(w) / gun(g) : ")
player_turn = Swg_game(computer, player)
print("\n")

print("************************************")

#printing the values selected by computer and the player
print(f"The value selected by the computer is : {computer}")
print(f"The value selected by the player is : {player}")
print("\n")

print("*************************************")

#declaring the result
if player_turn == None:
    print ("the game ia a tie!!")
elif player_turn:
    print("congratulations,you win!!!")
else:
    print("you lose!!\ntry again")

print("\n")
```
